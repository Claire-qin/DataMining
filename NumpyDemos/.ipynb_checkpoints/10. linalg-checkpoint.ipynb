介绍了numpy中的线性代数处理模块linalg

```python
import numpy as np
```

```python
# 两个数组的点积（内积）
# 对于二维向量，其等效于矩阵乘法。对于一维数组，它是向量的内积。对于 N 维数组，它是 a 的最后一个轴上的和与 b 的倒数 第二个轴的乘积。
a = np.array([[1,2], [3,4]])
b = np.array([[11,12], [13,14]])
np.dot(a,b)
```

```python
# 两个向量的点积
np.vdot(a, b)
```

```python
# 一维数组的向量内积。对于更高的维度，它返回最后一个轴上的乘积和。
np.inner(np.array([1,2,3]), np.array([0,1,2]))
```

```python
# 二维的维度，它返回最后一个轴上的乘积和
a = np.array([[1,2], [3,4]])
b = np.array([[11,12], [13,14]])
np.inner(a,b)      
# 1*11+2*12， 1*13+2*14， 
# 3*11+4*12， 3*13+4*14
```
