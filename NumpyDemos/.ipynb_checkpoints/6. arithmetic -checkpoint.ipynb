介绍了numpy中的算数运算

进行算数运算的数组必须有相同的shape或者符号广播机制

```python
import numpy as np
```

```python
a = np.arange(9).reshape(3,3)
b = np.array([10, 10, 10])
print(a)
print(b)
```

```python
# 相加
np.add(a,b)   # 广播
```

```python
# 相减
np.subtract(a, b)
```

```python
# 相乘
np.multiply(a,b)
```

```python
# 相除
np.divide(a, b)
```

```python
# 逐元素求倒数
a = np.array([0.1, 0.2, 0.25])
np.reciprocal(a)
```

```python
# 逐元素求指数，第一个为底，第二个为幂次
a = np.array([1,2,3])
b = np.array([1,2,3])
np.power(a,b)
```

```python
# 逐元素求余数
a = np.array([5,7,9])
b = np.array([1,2,3])
np.mod(a, b)
```

```python
# 比较两个同维数组间的最大值
x = np.random.randn(8)
y = np.random.randn(8)
np.maximum(x, y)
```

```python
# 返回数组逐元素的整数和小数部分
np.modf(x)
```

```python
# 两个数组坐标的网格化
xs, ys = np.meshgrid(np.linspace(0,1, 100), np.linspace(0,1, 100))
xs
```

```python
import matplotlib.pyplot as plt
z = np.sqrt(xs**2+ys**2)
z
```

```python
plt.imshow(z, cmap=plt.cm.gray); plt.colorbar()
plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
plt.plot()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(xs, ys, z, rstride=1, cstride=1, cmap=plt.cm.jet)
plt.show()
```
