介绍了利用Numpy进行FFT

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 30) 
wave = np.cos(x)
plt.plot(x, wave)
```

```python
# 对波形信号进行fft变换
transformed = np.fft.fft(wave)
plt.plot(transformed)
```

```python
# 在对频域信号进行逆fft变换，用于恢复信号
restored = np.fft.ifft(transformed)
np.abs(restored - wave).sum()
```
