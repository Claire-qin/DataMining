介绍了numpy中的线性代数处理模块linalg

```python
import numpy as np
```

#### 1. np.dot

```python
# 两个数组的点积（内积）
# 对于二维向量，其等效于矩阵乘法。对于一维数组，它是向量的内积。对于 N 维数组，它是 a 的最后一个轴上的和与 b 的倒数 第二个轴的乘积。
a = np.array([[1,2], [3,4]])
b = np.array([[11,12], [13,14]])
np.dot(a,b)
```

#### 2. np.vdot

```python
# 两个向量的点积
np.vdot(a, b)
```

#### 3. np.inner

```python
# 一维数组的向量内积。对于更高的维度，它返回最后一个轴上的乘积和。
np.inner(np.array([1,2,3]), np.array([0,1,2]))
```

```python
# 二维的维度，它返回最后一个轴上的乘积和
a = np.array([[1,2], [3,4]])
b = np.array([[11,12], [13,14]])
np.inner(a,b)      
# 1*11+2*12， 1*13+2*14， 
# 3*11+4*12， 3*13+4*14
```

#### 4. np.matmul

```python
# np.matmul() 返回数组的矩阵乘积
# 如果任一参数的维数大于2，则将其视为存在于最后两个索引的矩阵的栈，并进行相应广播。
# 如果任一参数是一维数组，则通过在其维度上附加 1 来将其提升为矩阵，并在乘法之后被去除。
```

```python
# 矩阵相乘
a = [[1,0],[0,1]]
b = [[4,1],[2,2]]
np.matmul(a, b)
```

```python
# 二维和一维运算
a = [[1,0], [0,1]]
b = [1,2]
print(np.matmul(a,b))    # 两次b升维的维度不一样
print(np.matmul(b,a))
```

```python
# 维数大于2的数组
a = np.arange(8).reshape(2,2,2)
b = np.arange(4).reshape(2,2)
print(a)
print("-"*10)
print(b)
print("-"*10)
print(np.matmul(a,b))
```
