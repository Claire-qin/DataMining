介绍了ndarray数据类型和常见的属性

```python
import numpy as np
```

#### dtype：可构造结构化的数据

```python
# 定义了含有三个字段name、age、scores的结构化数据student
# 内建类型都有一的字符代码：'b' ：布尔值；'i' ：符号整数；'u' ：无符号整数；'f' ：浮点， 'c' ：复数浮点
# 'm' ：时间间隔， 'M' ：日期时间， 'O' ：Python 对象， 'S', 'a' ：字节串， 'S', 'a' ：字节串， 'V' ：原始数据（ void ）
student = np.dtype([('name', 'S20'), ('age', 'i1'), ('scores', 'f4')])
student
```

```python
a = np.array([('Bob', 10, 88), ('Tom', 9, 89), ('Alex', 10, 95)])
a
```

```python
# 查看字段
student.fields
```
