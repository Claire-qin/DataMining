利用opencv对图片进行几何变换

图像的几何变换包括扩展缩放、平移、翻转、旋转、透视等<br>
几何变换的本质为向量矩阵的线性变换或放射变换，通过一个变换矩阵作用于原图得到变换后的图像<br>
opencv中提供了cv2.warpAffine(放射类变换，2*3的变换矩阵)和cv2.warpPerspective（透视类变换，3*3的变换矩阵）

图片（以RGB通道为例）的shape：(高度、宽度、通道数)，高度对应row，宽度对应col<br>
而从几何变换的坐标而言，X轴为宽度，Y轴为高度，所以变换矩阵的顺序为col——>row

对于图像而言，宽度方向是x，高度方向是y。原点的位置不是左下角而是左上角，y的方向也不是向上，而是向下。

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 扩展缩放

缩放时可以指定缩放后的目标图像大小，也可指定各维（宽度或高度方向）的缩放因子<br>
扩展缩放有不同的插值方法，对于缩放一般使用cv2.INTER_AREA，对于扩展，可使用cv2.INTER_CUBIC（慢)和cv2.INTER_LINEAR<br>
若无指定，则默认扩展缩放均按 cv2.INTER_LINEAR

```python
img = cv2.imread('./images/sai.jpeg') 
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# 此处指定各维度缩放因子，也可以直接指定目标图像大小
# 若指定dsize=()，则需注意元组顺序为(width, height)，而通过img.shape[0:2]得到的为(height, width)
# 其它变换函数中的dsize参数也是如此设计的
img_resize = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

print(img.shape)
print(img_resize.shape)
```

#### 平移

平移变换矩阵为：$M=\begin{bmatrix} 1&0&\Delta x\\0&1&\Delta y \end{bmatrix}$

```python
# 定义平移矩阵
M = np.float32([[1, 0, 100], [0, 1, 50]])   # 必须为np.float32或np.float64
row, col = img.shape[:2]
img_trans = cv2.warpAffine(img, M, (col, row))  # 这边要交换col和row
plt.imshow(img_trans)
plt.axis('off')
plt.show()
```

#### 旋转变换

如果是在原点处进行旋转（即线性变换），变换矩阵为$M=\begin{bmatrix}cos\theta&-sin\theta\\sin\theta&cos\theta\end{bmatrix}$

opencv提供了在任意位置进行旋转+缩放的功能：

式子中scale表示旋转后的缩放因子，center表示旋转中心，$\theta$为旋转角度

```python
# 生成旋转变换矩阵,getRotationMatrix2D(center, angle, scale),参数分别表示旋转中心、旋转角度和旋转后的缩放因子
row, col = img.shape[0:2]
M = cv2.getRotationMatrix2D((col/2, row/2), 45, 0.6)
M
```

```python
img_rotate = cv2.warpAffine(img, M, dsize=(col, row))
plt.imshow(img_rotate)
plt.axis('off')
plt.show()
```

#### 翻转变换

cv2.flip(src, flipCode[, dst]) → dst1 <br>
flipCode决定翻转模式：<br>
flipCode==0， 垂直翻转（沿X轴）<br>
flipCode>0， 水平翻转（沿Y轴）<br>
flipCode<0, 水平垂直翻转（先沿X轴翻转，再沿Y轴翻转，等价于旋转180°）<br>

```python
fig, ax = plt.subplots(2, 2)
img_flip_h = cv2.flip(img, flipCode=1)
img_filp_v = cv2.flip(img, flipCode=0)
img_filp_hv = cv2.flip(img, flipCode=-1)
ax[0][0].imshow(img)
ax[0][0].set_title("原图")
ax[0][1].imshow(img_flip_h)
ax[0][1].set_title("水平翻转")
ax[1][0].imshow(img_filp_v)
ax[1][0].set_title("垂直翻转")
ax[1][1].imshow(img_filp_hv)
ax[1][1].set_title("水平垂直翻转")

plt.show()
```

#### 放射变换

在仿射变换中，原图中所有的平行线在结果图像中同样平行。<br>
图像的仿射变换是深度学习预处理中常到的功能, 是上述缩放（sacle）、平移（translate）、旋转（rotate）和翻转（flip）四种变换的组合
从数学上讲，放射变换矩阵$M=\begin{bmatrix}A&B\end{bmatrix}=\begin{bmatrix}a_{11}&a_{12}&b_1\\a_{21}&a_{22}&b_2\end{bmatrix}$，其中$A$为线性变换项，$B$为平移项

为了创建这个矩阵我们需要从原图像中找到三个点以及他们在输出图像中的位置。然后cv2.getAﬃneTransform 会创建一个 2x3 的矩阵，最后这个矩阵会被传给函数 cv2.warpAﬃne。

```python
pts1=np.float32([[50,50],[200,50],[50,200]]) 
pts2=np.float32([[10,100],[200,50],[100,250]])
row, col = img.shape[:2]
M = cv2.getAffineTransform(pts1, pts2)
img_affine = cv2.warpAffine(img, M, (col, row))
plt.imshow(img_affine)
plt.axis('off')
plt.show()
```

#### 透视变换

对于视角变换， 我们需要一个 3x3 变换矩阵。 在变换前后直线还是直线。 要构建这个变换矩阵， 你需要在输入图像上找 4 个点， 以及他们在输出图 像上对应的位置。 这四个点中的任意三个都不能共线。 这个变换矩阵可以有 函数 cv2.getPerspectiveTransform() 构建。 然后把这个矩阵传给函数 cv2.warpPerspective。

```python
img2 = cv2.imread('./images/box.png')
plt.imshow(img2)
plt.axis('off')
plt.show()
```
