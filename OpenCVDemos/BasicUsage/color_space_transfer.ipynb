利用opencv进行图像颜色空间的变换

```python
import cv2
```

#### 颜色空间

opencv支持150+种颜色空间的变换，常用的有BGR——>RGB、BGR——>HSV(色彩/色度、饱和度和亮度)、BGR——>Gray（灰度）

在 OpenCV 的 HSV 格式中，H（色彩/色度）的取值范围是 [0，179]， S（饱和度）的取值范围 [0，255]，V（亮度）的取值范围 [0，255]。但是不 同的软件使用的值可能不同。所以当你需要拿 OpenCV 的 HSV 值与其他软 件的 HSV 值进行对比时，一定要记得归一化。

```python
[i for i in dir(cv2) if i.startswith('COLOR_')]
```

#### 找到RGB通道内的某个颜色对应的HSV通道对应值

```python
# 这里使用一维数组是不对的
import numpy as np
green = np.uint8([0,255,0])
hsv_green=cv2.cvtColor(green,cv2.COLOR_BGR2HSV)
```

```python
# 要用三维数组去表示颜色
green=np.uint8([[[0,255,0]]]) 
hsv_green=cv2.cvtColor(green,cv2.COLOR_BGR2HSV)
print(hsv_green)
```
