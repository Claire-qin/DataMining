介绍了pandas里面的index对象基本操作

常见的index类型有Index(基类)、Int64Index、MultiIndex(多层索引)、DatetimeIndex（时间索引）和PeriodIndex

```python
import pandas as pd
import numpy as np
```

```python
A = pd.DataFrame(np.arange(12).reshape(4,3))
B = pd.DataFrame(np.arange(9).reshape(3,3))
```

```python
# 查看index
A.index
```

```python
# 两个index的连接
A.index.append(B.index)
```

```python
# 两个index的差集
A.index.difference(B.index)
```

```python
# 两个index的并集
A.index.union(B.index)
```

```python
# 两个index的交集
A.index.intersection(B.index)
```

```python
# 查看index是否在某集合中
A.index.isin([1,3,5])
```

```python
# 删除索引，注意参数为位置
A.index.delete(2)
```

```python
# 删除指定的索引, 注意参数为索引名称
A.index.drop(2)
```

```python
# 在指定位置插入索引
A.index.insert(2, 100)
```

```python
# 查看所有index是否后一个均大于前一个
A.index.is_monotonic
```

```python
# 查看所有index是否均唯一
A.index.is_unique
```

```python
# 查看唯一的index
A.index.unique
```

```python
# index有一个name的属性
A.index.name
```

```python
A.index.name = "name"
A.index.name
```
