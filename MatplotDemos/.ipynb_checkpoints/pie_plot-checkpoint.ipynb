利用matplotlib画饼图

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.figure(1, figsize=(8,8))
# ax = plt.axes([0.1, 0.1, 0.8, 0.8])    # 坐标轴范围

labels = ["spring", "summer", "autumn", "winter"]
x = [15, 30, 45, 10]

explode = (0.1, 0.1, 0.1, 0.1)    # 分裂距离
plt.pie(x, explode=explode, autopct='%1.1f%%', labels=labels, startangle=67)
# explode分裂情况，startangle起始角度
plt.title('Rainy days by season')
plt.show()
```
