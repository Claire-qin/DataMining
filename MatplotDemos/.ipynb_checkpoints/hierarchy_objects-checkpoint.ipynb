介绍了matplot的基本图像元素和的之间的层次关系<br>
plot、legend等命令生成对应元素时，会返回对应的元素，通过捕获返回对象可以进行进一步的设置

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = range(1, 5)
y = [13, 24, 6, 19]
```

```python
fig = plt.figure(figsize=(6, 4))   # fig对象
ax = fig.add_subplot(1,1,1)   # 在fig中添加axes
ax.plot(x, y)
plt.show()
```

### axes对象

#### 普通用法

```python
# axes提供了设置title、label的方法
fig = plt.figure(figsize=(6, 4))   # fig对象
ax = fig.add_subplot(1,1,1)   # 在fig中添加axes
ax.set_title("示例", fontdict={"size":16})
ax.set_xlabel("X轴", fontdict={"size":15})
ax.set_ylabel("Y轴", fontdict={"size":15})
ax.plot(x, y)
plt.show()
```

```python
# axes的set_xlim和set_ylim给出了作图取值的范围
fig = plt.figure(figsize=(6, 4))   # fig对象
ax = fig.add_subplot(1,1,1)   # 在fig中添加axes
ax.set_title("示例", fontdict={"size":16})
ax.set_xlim(0, 5)
ax.set_ylim(0, 30)
ax.set_xlabel("X轴", fontdict={"size":15})
ax.set_ylabel("Y轴", fontdict={"size":15})
ax.plot(x, y)
plt.show()
```

```python
# ax的tick_params可以调整tick的相关参数, set_yticks和set_yticklabels
fig = plt.figure(figsize=(6, 4))   # fig对象
ax = fig.add_subplot(1,1,1)   # 在fig中添加axes
ax.set_title("示例", fontdict={"size":16})
ax.set_xlim(0, 5)
ax.set_ylim(0, 30)
ax.set_xlabel("X轴", fontdict={"size":15})
ax.set_ylabel("Y轴", fontdict={"size":15})
ax.set_xticks([1, 1.5, 2, 2.5, 3, 3.5, 4])   # 在指定数值处设置刻度
ax.set_xticklabels(['a','b','c','d','e','f','g'])  # 刻度处对应的label
ax.tick_params(labelsize=10)   # tick和ticklabel的相应参数
ax.plot(x, y)
plt.show()
```

#### 通过set一次性设置

```python
fig = plt.figure(figsize=(6, 4))   # fig对象
ax = fig.add_subplot(1,1,1)   # 在fig中添加axes
ax.set(title="示例",xlim=[0, 5],ylim=[0,30],xlabel="X轴",ylabel="Y轴")
ax.plot(x, y)
plt.show()
```

### axis对象

```python
# xaxis和yaxis获取坐标轴对象从而进行get或set操作
ax.xaxis.set_visible(False)
```

```python
# 获取xaxis的label(即xlabel), 进行get或set操作
ax.xaxis.label.set_text()
```

```python
ax.xaxis.get_ticklocs()    # 获取tick的位置
```

```python
for i in ax.xaxis.get_ticklabels():   # 获取xticklabel的信息
    print(i)
```

### legend对象

```python
fig = plt.figure(figsize=(6, 4))   # fig对象
ax = fig.add_subplot(1,1,1)   # 在fig中添加axes
ax.plot(x, y, label='demo')   # label设置legend的handle
leg = ax.legend(loc='best',shadow=True, fancybox=True)   # 配置参数，返回legend对象
leg.get_frame().set_alpha(0.5)   # 获取legend的frame并进行参数设置
plt.show()
```
