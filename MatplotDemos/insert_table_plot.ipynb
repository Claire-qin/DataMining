在图中插入表格

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.figure(figsize=(8,8))
ax = plt.gca()
y = np.random.randn(9)

col_labels = ['c1', 'c2', 'c3']
row_labels =['r1', 'r2', 'r3']
table_values = [[1,2,3], [4,5,6], [7,8,9]]
row_colors = ['red', 'gold', 'green']

my_table = plt.table(cellText=table_values, colWidths=[0.1] * 3, rowLabels=row_labels,
                    colLabels=col_labels, rowColours=row_colors, loc='best')
plt.plot(y)
plt.show()
```
