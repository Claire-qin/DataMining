利用matplotlib画三维bar图, 在mpl_toolkits.mplot3d中提供了三维图形的相关功能

```python
import random
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as mdats
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
```

```python
mpl.rcParams['font.size'] = 10
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')   # 通过设置projection='3d'支持3D图

for z in [2011, 2012, 2013, 2014]:
    xs = np.arange(1, 13)
    ys = 1000 * np.random.rand(12)
    color = plt.cm.Set2(random.choice(np.arange(plt.cm.Set2.N)))
    ax.bar(xs, ys, zs=z, zdir='y', color=color, alpha=0.8)   # zdir指定z的方向
    
ax.xaxis.set_major_locator(mpl.ticker.FixedLocator(xs))
ax.xaxis.set_major_locator(mpl.ticker.FixedLocator(ys))
ax.set_xlabel('Month')
ax.set_ylabel('Year')
ax.set_ylabel('Sales Net [usd]')

plt.show()
```
