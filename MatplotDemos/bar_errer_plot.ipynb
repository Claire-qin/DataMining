利用matplotlib画误差条状图

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# 生成数据
x = np.arange(1, 10 ,1)
y = np.log(x)

xe = 0.1 * np.abs(np.random.randn(len(x)))    # 误差界限
```

```python
# 绘制误差条状图
fig = plt.figure(figsize=(10, 8))
plt.bar(x, y ,yerr=xe, width=0.4, align='center', edgecolor='r', hatch='/', color='cyan', label='error')
plt.xlabel("x value", fontsize=12)
plt.ylabel("y value", fontsize=12)
plt.legend(loc='upper left')
plt.show()
```
