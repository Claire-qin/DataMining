利用matplotlib画散点图

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# 生成数据
x = np.random.randn(1000)

y1 = np.random.randn(len(x))
y2 = 1.2 + np.exp(x)
```

```python
# 绘制scatter
fig = plt.figure(figsize=(10,10))
ax1 = plt.subplot(121)
plt.scatter(x, y1, color='indigo', alpha=0.3, edgecolors='white', label='不相关')
plt.grid()
plt.legend()

ax2 = plt.subplot(122)
plt.scatter(x, y2, color='green', alpha=0.3, edgecolors='grey', label='相关')
plt.grid()
plt.legend()

plt.show()
```
