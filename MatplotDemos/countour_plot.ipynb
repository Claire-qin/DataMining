利用matplotlib画等高线图

matplotlib提供了contour(等高线)和contourf(填充的等高线)两种方式
在做等高线图时要注意：
（1）添加标签，以使等高线有意义，一般采用clabel或colormaps（如果输出媒体允许使用颜色，首选）添加标签；
（2）等值线不要太密集，以防止不清晰

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import warnings
%matplotlib inline
```

```python
# 生成数据
def process_signal(x, y):
    return (1-(x**2 + y**2)) * np.exp(-y**3/3)

x = np.arange(-1.5, 1.5, 0.1)
y = np.arange(-1.5, 1.5, 0.1)

X, Y = np.meshgrid(x, y)
Z = process_signal(X, Y)
```

```python
# 等高线的值
N = np.arange(-1.5, 1.5, 0.3)
```

```python
# 绘制等高线图
CS = plt.contour(Z, N, linewidth=2, camp=mpl.cm.jet)
# 添加标签
plt.clabel(CS, inline=True, fmt='%1.1f', fontsize=10)
# colorbar
plt.colorbar(CS)

plt.title("$z=(1-x^2+y^2)e^{-(y^3)/3}$")
plt.show()
```
