介绍了python中的位运算

#### 1. 左移运算<<

左移1位相当于乘2。各二进位全部左移n位,高位丢弃,低位补0。<p>
左移可能会改变一个数的正负性

```python
5 << 2    # 左移动两位，相当于乘4
```

#### 2. 右移运算>>

右移1位相当于除以2。各二进位全部右移n位,保持符号位不变。<p>
移出的位删掉,移进的位补符号位,右移不会改变一个数的符号

```python
6 >> 1    # 右移一位，相当于除2
```

```python
5 >> 1    # 若不能整除，相当于取整
```

#### 3. 按位与&

只有对应的两个二进位均为1时,结果位才为1,否则为0 <p>
利用按位与可以将任意二进制数的最后一位变为0,即就是X&0

```python
3 & 5    # 011&101——>001=1
```

```python
-1 & -2   # 11111111&11111110(补码相加)——>11111110(补码)——>10000010（原码）=-2
```

```python
-2 & 6   # 11111110&00000110(补码相加)——>00000110(补码)——>00000110（原码）
```
